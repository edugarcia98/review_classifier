Funções para extração de marcas e modelos de Smart TVs e Smartphones

```python
def extract_products(site_id: str, q: str, limit: int) -> List[str]:
    params = {"site_id": site_id, "q": q, "limit": limit}

    response = requests.get(
        "https://api.mercadolibre.com/products/search",
        params=params,
    )
    
    brands = {item["attributes"][0]["value_name"] for item in response.json()["results"]}
    
    return brands
```

```python
def extract_brands_and_lines(site_id: str, q: str, brand_list: List[str], limit: int) -> List[str]:
    brands_and_lines = []
    
    for brand in brand_list:
        params = {"site_id": site_id, "q": f"{q} {brand}", "limit": limit}

        response = requests.get(
            "https://api.mercadolibre.com/products/search",
            params=params,
        )

        lines = {item["attributes"][1]["value_name"] for item in response.json()["results"]}

        values = [f"{brand} {line}" for line in lines]

        brands_and_lines.extend(values)
    
    return brands_and_lines
```

### Extração de Dados

Criação de funções para extração produtos de determinada marca e seus respectivos reviews

```python
ERROR_QUERIES = []

def load_products_by_query(q: str) -> Optional[List[dict]]:
    print(f"Searching {q}\n")
    
    params = {"site_id": "MLB", "q": q, "limit": 100, "status": "active"}
    
    try:
        response = requests.get(
            "https://api.mercadolibre.com/products/search",
            params=params,
        )
        
        results = response.json()["results"]
    except Exception as exc:
        print(f"Unable to load products from {q}. Reason: {exc}.\n")
        
        ERROR_QUERIES.append(q)
        
        return []
    
    return results
```

```python
ERROR_PRODUCT_RELEVANT_DATA = []

def load_product_relevant_data(item: dict) -> Optional[dict]:
    try:
        product_data = {
            "product_id": item["id"],
            "product_name": item["name"],
            "product_brand": item["attributes"][0]["value_name"],
            "product_line": item["attributes"][1]["value_name"],
        }
    except (KeyError, IndexError) as exc:
        print(f"Missing item info. {exc}.\n")
        
        ERROR_PRODUCT_RELEVANT_DATA.append(item)
        
        return None
    
    return product_data
```

```python
ERROR_PRODUCT_REVIEWS = []

def load_product_reviews(product_id: str, name: str) -> Optional[List[dict]]:
    params = {"catalog_product_id": product_id, "limit": 100}
    
    try:
        response = requests.get(
            "https://api.mercadolibre.com/reviews/item/MLB1002",
            params=params,
        )
        
        reviews = response.json()["reviews"]
    except Exception as exc:
        print(
            f"Unable to load product {name} {product_id} reviews. "
            f"Reason: {exc}."
        )
        
        ERROR_PRODUCT_REVIEWS.append(product_id)
        
        return None
    
    if not reviews:
        print(f"Product {name} {product_id} has no reviews.")
        
        return None
    
    print(
        f"Product {name} {product_id} reviews successfully loaded. ({len(reviews)} reviews)"
    )
    
    return reviews
```

### Dataframe - Smart TVs

Criando dataframe de reviews de Smart TVs

```python
cols = [
    "product_id",
    "product_name",
    "product_brand",
    "product_line",
    "review_date",
    "review_text",
    "review_rate",
    "review_likes",
    "review_dislikes",
    "review_relevance",
    "client_buying_date",
]

df = pd.DataFrame(columns=cols)

idx = 0

for item in brands_and_lines:
    q = f"Smart TV {item}"
    
    results = load_products_by_query(q)
    
    for item in results:
        if not (product_data := load_product_relevant_data(item)):
            continue
        
        if not (
            reviews := load_product_reviews(
                product_data["product_id"], product_data["product_name"],
            )
        ):
            continue
        
        for review in reviews:
            title = review.get("title", "")
            content = review.get("content", "")
            
            review_data = {
                "review_date": review.get("date_created"),
                "review_text": f"{title} {content}".replace("\n", ""),
                "review_rate": review.get("rate"),
                "review_likes": review.get("likes"),
                "review_dislikes": review.get("dislikes"),
                "review_relevance": review.get("relevance"),
                "client_buying_date": review.get("buying_date")
            }
            review_data.update(product_data)
            
            item = pd.DataFrame(review_data, index=[idx])
            df = pd.concat([df, item])
            
            idx += 1
    
    print("\n")
        
```
